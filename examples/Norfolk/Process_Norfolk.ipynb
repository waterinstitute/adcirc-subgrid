### Time Difference
The following code plots the times collected from running subgrid-adcirc and storing the runtimes.

````python
resolutions = [10, 30, 100]
runtimes = [runtime10, runtime30, runtime100]
````

````python
plt.figure(figsize=(6, 4))
plt.plot(resolutions, runtimes, marker='o', linestyle='-', color='teal')
plt.xscale("log")
plt.yscale("log")
plt.xticks(resolutions, labels=["10", "30", "100"])
ax = plt.gca()
ax.yaxis.set_major_formatter(NullFormatter())  # Remove all tick labels
ax.yaxis.set_minor_formatter(NullFormatter())
plt.yticks(runtimes, labels=["{} s".format(x) for x in runtimes])

plt.xlabel('DEM Resolution (meters)')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime vs DEM Resolution')
plt.grid(True)

plt.show()
````

````python
plt.figure(figsize=(7, 5))
bars = plt.bar(resolutions, runtimes, color=["#35b41f", "#d7ff0e", "#a02c60"])

# Calculate percentage improvements (relative to previous resolution)
percentage_improvements = [None]  # No improvement for the first
for i in range(1, len(runtimes)):
    prev = runtimes[i - 1]
    curr = runtimes[i]
    improvement = (prev - curr) / prev * 100
    percentage_improvements.append(improvement)


# Add labels and title
plt.xlabel('DEM Resolution (m)')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime vs DEM Resolution')
plt.xticks(resolutions)
plt.yscale("log")
#plt.ylim([40, 100])  # Adjusted to fully include runtimes
ax = plt.gca()
ax.yaxis.set_major_formatter(NullFormatter())  # Remove all tick labels
ax.yaxis.set_minor_formatter(NullFormatter())

# Add labels only at the data points (runtimes)
ax.set_yticks(runtimes)
ax.set_yticklabels([f"{y} s" for y in runtimes])

plt.grid(axis='y', linestyle='--', alpha=0.7)

## annotate bars with the exact timings and the percentage decrease from the previous step
for i, (bar, runtime) in enumerate(zip(bars, runtimes)):
    yval = bar.get_height()
    label = f'{runtime:.1f}s'
    if percentage_improvements[i] is not None:
        label += f'\n↓ {percentage_improvements[i]:.1f}%'
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, label,
             ha='center', va='center_baseline', fontsize=9)

plt.tight_layout()
plt.show()
````
